```python
Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.
ans:-
Matplotlib is a popular data visualization library in Python that provides a wide range of tools to create various types of static, animated, and interactive plots.
Matplotlib is used to create high-quality and customizable visualizations in Python, such as line plots, scatter plots, bar plots, histograms, heatmaps, and more.
It provides a lot of flexibility and control over the appearance and layout of the plots, allowing users to create customized visualizations that suit their specific needs.

Matplotlib is used to create high-quality and customizable visualizations in Python, such as line plots, scatter plots, bar plots, histograms, heatmaps, and more. 
It provides a lot of flexibility and control over the appearance and layout of the plots, allowing users to create customized visualizations that suit their specific needs.

Here are five common plots that can be plotted using the Pyplot module of Matplotlib:

1) Line plot: A line plot is a basic plot that displays data as a series of points connected by straight lines. It is commonly used to visualize trends over time or to show the relationship between two variables.

2) Scatter plot: A scatter plot is a plot that displays data as a collection of points. It is often used to show the relationship between two variables or to identify any patterns or outliers in the data.

3) Bar plot: A bar plot is a plot that displays data as bars. It is commonly used to compare the values of different categories or to show changes over time.

4) Histogram: A histogram is a plot that displays the distribution of a numerical variable. It is commonly used to show the shape and spread of a data distribution.

5) Heatmap: A heatmap is a plot that displays data as a color-coded matrix. It is commonly used to visualize correlations or patterns in large datasets.
```

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
ans:
The subplot() function in Matplotlib is used to create multiple plots within a single figure. 
It is particularly useful when you want to compare multiple plots side by side or when you want to create a complex layout of plots in a single figure.

The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current plot. The index starts at 1 and increases 
from left to right and top to bottom.

import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
# Create the subplots
fig, axs = plt.subplots(nrows=2, ncols=2)
# Plot line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a title to the figure
fig.suptitle('Four Line Plots')

# Display the plot
plt.show()
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
ans:
A bar plot is a plot that displays data as bars. It is commonly used to compare the values of different categories or to show changes over time.
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.title("Company Profit")
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.show()

# Create a horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profit")
plt.xlabel("Profit ($)")
plt.ylabel("Company")
plt.show()
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
ans:
A box plot is a type of plot that displays the distribution of data based on their quartiles, outliers, and the range of the data. 
The plot is composed of a box that represents the interquartile range (IQR), which is the distance between the first and third quartiles 
(25th and 75th percentiles) of the data. The box also contains a horizontal line that represents the median of the data. Additionally, 
there are whiskers that extend from the box to represent the range of the data, excluding any outliers, which are plotted as individual points.
import numpy as np
import matplotlib.pyplot as plt

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create a box plot
plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Values')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
