```python
# Q1. How can you create a Bokeh plot using Python code?
# ans: Bokeh is a Python library for creating interactive and dynamic data visualizations in web browsers. Heres a simple example of how to create a Bokeh plot using Python code:
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show

# create a figure object
p = figure(title="My Bokeh Plot", x_axis_label="x-axis", y_axis_label="y-axis")

# add a scatter plot with some data
p.scatter([1, 2, 3, 4, 5], [2, 5, 4, 6, 7], size=10, color="blue")

# specify output file
output_file("bokeh_plot.html")

# show the plot
show(p)

```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
# ans: In Bokeh, glyphs are visual shapes or markers that can be used to represent data points in a plot. 
# They can be customized in terms of size, color, and other properties to convey information about the data being plotted. 
# Bokeh provides a wide range of glyph types, including circles, squares, triangles, lines, and more.
# from bokeh.plotting import figure, output_file, show

# create a figure object
p = figure(title="My Bokeh Plot", x_axis_label="x-axis", y_axis_label="y-axis")

# add a circle glyph with some data
p.circle([1, 2, 3, 4, 5], [2, 5, 4, 6, 7], size=20, fill_color="red", line_color="black")

# specify output file
output_file("bokeh_plot.html")

# show the plot
show(p)
```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
# ans:Bokeh provides several options for customizing the appearance of a plot, including the axes, title, and legend. Here are some examples of how to do this:
#1. Customizing the title:
from bokeh.plotting import figure, output_file, show

# create a figure object
p = figure(title="My Bokeh Plot", x_axis_label="x-axis", y_axis_label="y-axis")

# customize the title
p.title.text_color = "blue"
p.title.text_font_size = "20pt"
p.title.align = "center"
# customize the x-axis
p.xaxis.axis_label_text_color = "red"
p.xaxis.axis_label_text_font_size = "14pt"
p.xaxis.major_label_text_color = "green"
p.xaxis.major_label_text_font_size = "12pt"

# customize the y-axis
p.yaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_font_size = "14pt"
p.yaxis.major_label_text_color = "purple"
p.yaxis.major_label_text_font_size = "12pt"

# customize the legend
p.legend.title = "Legend"
p.legend.location = "top_left"
p.legend.label_text_font_size = "12pt"

# add a scatter plot with some data
p.scatter([1, 2, 3, 4, 5], [2, 5, 4, 6, 7], size=10, color="blue")

# specify output file
output_file("bokeh_plot.html")

# show the plot
show(p)
```

```python
#@title Default title text
# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
# real time?
# A Bokeh server is a way to create interactive Bokeh plots that can be updated in real time. 
# A Bokeh server is a Python application that runs on a web server and communicates with a client web browser to display and update the plot.
# To create a Bokeh server application, you need to define a curdoc() function that creates a Bokeh plot, and 
# then run the application using the bokeh serve command. The curdoc() function is a special function that represents the current Bokeh document, 
# which is the top-level container for all the Bokeh objects that will be displayed in the plot.
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.layouts import row
from bokeh.models.widgets import Slider

def create_plot():
    # create a figure object
    p = figure(x_range=(0, 10), y_range=(0, 10))
    
    # add a scatter plot with some data
    source = ColumnDataSource(data=dict(x=[1, 2, 3], y=[2, 5, 4]))
    p.scatter(x='x', y='y', source=source, size=10)
    
    # add a slider widget
    slider = Slider(start=0, end=10, value=5, step=0.1, title="Slider")
    
    # define a callback function for the slider
    def callback(attr, old, new):
        # update the data source based on the slider value
        source.data = dict(x=[1, 2, 3], y=[new, new+3, new+2])
    
    # attach the callback function to the slider value
    slider.on_change('value', callback)
    
    # return the plot and slider widgets as a layout
    return row(p, slider)

doc = curdoc()

# call the create_plot function to create the initial plot and layout
layout = create_plot()

# add the layout to the document
doc.add_root(layout)

# bokeh serve --show myapp.py
```
