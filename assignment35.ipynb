```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.
ans:
A t-test and a z-test are statistical tests used to compare means of two samples and make inferences about the population. The main difference between the 
two tests is the underlying assumptions about the population variance.

A z-test is used when the population variance is known, or when the sample size is large enough to assume a normal distribution of the sample means. 
For example, suppose we want to compare the heights of two populations: men and women. We know the population variance of men and women, and the sample 
size is large enough to assume a normal distribution of the sample means. In this case, we can use a z-test to compare the means of the two samples and make 
inferences about the population.

On the other hand, a t-test is used when the population variance is unknown, or when the sample size is small. A t-test uses a t-distribution to estimate
the population variance based on the sample variance. For example, suppose we want to compare the average weight of two groups of mice, but we do not know 
the population variance of the weights of mice. In this case, we can use a t-test to compare the means of the two samples and make inferences about the 
population.
```

```python
Q2: Differentiate between one-tailed and two-tailed tests.
ans:
In hypothesis testing, the alternative hypothesis is the statement that the researcher is trying to support, and it is either one-tailed or two-tailed. 
The main difference between one-tailed and two-tailed tests is the directionality of the alternative hypothesis.

In a one-tailed test, the alternative hypothesis is directional, meaning that it specifies a specific direction for the parameter being tested. For example, 
the alternative hypothesis might be that the mean of Group A is greater than the mean of Group B. The researcher is interested in detecting a difference in 
one direction only, either higher or lower, and not both.

In a two-tailed test, the alternative hypothesis is non-directional, meaning that it does not specify a specific direction for the parameter being tested.
Instead, it simply states that there is a difference between the groups being compared. For example, the alternative hypothesis might be that there is a 
difference in the mean scores of Group A and Group B. The researcher is interested in detecting a difference in either direction, either higher or lower.

The choice of one-tailed or two-tailed test depends on the research question and the direction of the effect that the researcher wants to detect. If the 
researcher is interested in detecting a difference in one direction only, a one-tailed test should be used. If the researcher is interested in detecting a
difference in either direction, a two-tailed test should be used.
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

In hypothesis testing, Type 1 and Type 2 errors are two possible errors that can occur.

Type 1 error, also known as a false positive, occurs when we reject the null hypothesis even though it is true. This error happens when we set our 
significance level (alpha) too low, leading us to conclude that there is a significant effect or relationship when there really is not. The probability of 
making a Type 1 error is denoted by alpha (α), and it is typically set at 0.05 or 0.01.

For example, suppose a company is testing a new drug and concludes that it is effective in treating a particular condition when it is not. This would be a 
Type 1 error because they rejected the null hypothesis (the drug is not effective) even though it was true.

Type 2 error, also known as a false negative, occurs when we fail to reject the null hypothesis even though it is false. This error happens when we set our 
significance level too high, leading us to conclude that there is no significant effect or relationship when there really is. The probability of making a 
Type 2 error is denoted by beta (β).

For example, suppose a company is testing a new drug and concludes that it is not effective in treating a particular condition when it actually is. 
This would be a Type 2 error because they failed to reject the null hypothesis (the drug is not effective) even though it was false.
```

```python
Q4: Explain Bayes's theorem with an example.
ans:
Bayes's theorem is a mathematical formula that allows us to update our beliefs about the likelihood of an event based on new evidence. It involves calculating the conditional probability of an event given some observed evidence.

The formula for Bayes's theorem is:

P(A|B) = P(B|A) x P(A) / P(B)

where:
P(A|B) = probability of event A given evidence B
P(B|A) = probability of evidence B given event A
P(A) = prior probability of event A (before considering the evidence)
P(B) = total probability of evidence B

An example scenario where Bayes theorem can be applied is in medical diagnosis. Lets say a certain disease affects 1 in 1000 
people in the population, and a certain test has a false positive rate of 5% and a false negative rate of 1%. If a person tests positive for 
the disease, what is the probability that they actually have the disease?

Using Bayes's theorem, we can calculate the probability as follows:

P(disease|positive test) = P(positive test|disease) x P(disease) / P(positive test)

P(positive test|disease) = 1 - false negative rate = 0.99
P(disease) = 0.001
P(positive test) = P(positive test|disease) x P(disease) + P(positive test|no disease) x P(no disease)
= 0.99 x 0.001 + 0.05 x 0.999
= 0.05094

P(disease|positive test) = 0.99 x 0.001 / 0.05094 = 0.0194

Therefore, the probability that a person who tests positive for the disease actually has the disease is only 1.94%. 
This shows that even with a relatively accurate test, the probability of a positive result being a true positive can still be quite low in a 
low prevalence population.
```

```python
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
ans:
A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence based on a sample 
statistic. It is used to estimate the population parameter when we cannot observe the entire population.

The formula for a confidence interval is:

Confidence interval = point estimate ± margin of error

where the point estimate is the sample statistic, and the margin of error is calculated using the standard error, the level of significance 
(or confidence level), and the sample size.

For example, suppose we want to estimate the average height of students in a school, and we take a random sample of 100 students. The sample mean 
height is 68 inches, and the sample standard deviation is 3 inches. If we want to calculate the 95% confidence interval for the population mean height, 
we can use the following formula:

Margin of error = Zα/2 * (σ/√n)

where Zα/2 is the critical value for the level of significance (α/2), σ is the population standard deviation (or the sample standard deviation if the 
                                                                                                              population standard deviation is unknown), 
and n is the sample size.

For a 95% confidence level, the level of significance (α) is 0.05, so α/2 is 0.025. The critical value for a 95% confidence level and a sample size of 100 
is 1.96 (based on the standard normal distribution table). Therefore, the margin of error is:

Margin of error = 1.96 * (3/√100) = 0.588

The 95% confidence interval is:

Confidence interval = 68 ± 0.588 = (67.412, 68.588)

This means that we can be 95% confident that the true population mean height falls between 67.412 inches and 68.588 inches.
```

```python
Q6. Use Bayes's Theorem to calculate the probability of an event occurring given prior knowledge of the
events probability and new evidence. Provide a sample problem and solution.
ans:
Suppose a certain disease affects 0.1% of the population. A test for the disease correctly identifies 95% of those who have the disease and 98% of those 
who do not have the disease. If a person tests positive, what is the probability that they actually have the disease?

Solution using Bayes' Theorem:

Let's use the following notation:

A: the event that a person has the disease
B: the event that a person tests positive
We want to find P(A|B), the probability that a person has the disease given that they test positive.

Using Bayes' Theorem, we have:

P(A|B) = P(B|A) * P(A) / P(B)

where:

P(B|A) = 0.95: the probability of testing positive given that the person has the disease
P(A) = 0.001: the prior probability of a person having the disease
P(B) = P(B|A) * P(A) + P(B|not A) * P(not A): the total probability of testing positive, where P(B|not A) = 1 - 0.98 = 0.02 and P(not A) = 1 - P(A)
Substituting in the values, we get:

P(A|B) = 0.95 * 0.001 / (0.95 * 0.001 + 0.02 * 0.999) = 0.045

Therefore, the probability that a person actually has the disease given that they test positive is only 4.5%. This is because the disease is rare and 
the test has a relatively high false positive rate.
```

```python
Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5. Interpret the results.
ans:
To calculate the 95% confidence interval, we need to use the formula:

CI = x̄ ± z*(σ/√n)

Where:
x̄ = sample mean (50)
σ = sample standard deviation (5)
n = sample size
z = z-score for the desired confidence level (95% confidence level corresponds to a z-score of 1.96)


Assuming a sample size of 30, we get:

CI = 50 ± 1.96*(5/√30)
CI = 50 ± 1.82

So, the 95% confidence interval for the population mean is (48.18, 51.82). This means that if we take multiple samples from the same population and calculate the confidence interval for each sample, we can expect 95% of those intervals to contain the true population mean.

In other words, we are 95% confident that the true population mean falls between 48.18 and 51.82.
```

```python
Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error?
Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
ans:
The margin of error in a confidence interval is a measure of the amount of error that can be expected in an estimate, given a specific level of confidence. 
It represents the range of values within which the true population parameter is expected to fall. The margin of error is calculated by multiplying the 
critical value (z-score or t-score) by the standard error of the estimate.

The margin of error is inversely proportional to the sample size. As the sample size increases, the margin of error decreases, and as the sample size 
decreases, the margin of error increases. This is because a larger sample size provides a more accurate representation of the population, resulting in a
more precise estimate.

For example, suppose a pollster wants to estimate the percentage of voters in a city who plan to vote for a particular candidate in an upcoming election.
If the pollster takes a sample of 100 voters and finds that 60% plan to vote for the candidate, with a margin of error of +/- 5%, this means that the true 
percentage of voters who plan to vote for the candidate is expected to fall between 55% and 65% with 95% confidence. However, if the pollster takes a 
sample of 1000 voters, the margin of error will be smaller, say +/- 1.6%, resulting in a more precise estimate of the true percentage of voters who plan to 
vote for the candidate.
```

```python
Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population
standard deviation of 5. Interpret the results.
ans:
To calculate the z-score, we can use the formula:

z = (x - μ) / σ

where x is the value of the data point, μ is the population mean, and σ is the population standard deviation.

Substituting the given values, we get:

z = (75 - 70) / 5

z = 1

The z-score is 1. This means that the data point is one standard deviation above the population mean.
```

```python
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test.
ans:
Null hypothesis: The weight loss drug is not significantly effective. The mean weight loss is equal to or less than 0 pounds.
Alternative hypothesis: The weight loss drug is significantly effective. The mean weight loss is greater than 0 pounds.
We will use a one-tailed t-test with a significance level of 0.05.

First, we calculate the t-statistic:

t = (sample mean - hypothesized mean) / (sample standard deviation / sqrt(sample size))
t = (6 - 0) / (2.5 / sqrt(50))
t = 15.10

Next, we find the degrees of freedom:

df = sample size - 1
df = 50 - 1
df = 49

Using a t-table or a t-distribution calculator with 49 degrees of freedom and a one-tailed test at a 0.05 significance level, we find the critical 
t-value to be 1.676.

Since our calculated t-statistic (15.10) is greater than the critical t-value (1.676), we can reject the null hypothesis and conclude that 
the weight loss drug is significantly effective at a 95% confidence level.
```

```python
Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.
Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82
with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a
significant difference in student performance using a t-test with a significance level of 0.01.
ans:

Let's set up the hypotheses:

Null hypothesis (H0): The means of sample A and sample B are equal.
Alternative hypothesis (Ha): The means of sample A and sample B are not equal.
We can use a significance level of 0.01, which means we are willing to reject the null hypothesis if the probability of obtaining the observed 
results by chance is less than 1%.

The formula for the two-sample t-test is:

t = (x1 - x2) / sqrt((s1^2/n1) + (s2^2/n2))

where:

x1 and x2 are the sample means
s1 and s2 are the sample standard deviations
n1 and n2 are the sample sizes
Let's calculate the t-value:

t = (85 - 82) / sqrt((6^2/50) + (5^2/50))
t = 3 / 1.128
t = 2.66

The degrees of freedom (df) can be calculated using the following formula:

df = (s1^2/n1 + s2^2/n2)^2 / ( (s1^2/n1)^2 / (n1-1) + (s2^2/n2)^2 / (n2-1) )

df = (6^2/50 + 5^2/50)^2 / ( (6^2/50)^2 / (50-1) + (5^2/50)^2 / (50-1) )
df = 94.39

Using a t-table with 94 degrees of freedom and a significance level of 0.01, we find the critical t-value to be ±2.62.

Since the calculated t-value (2.66) is greater than the critical t-value (±2.62), we can reject the null hypothesis and conclude that there is 
a significant difference in student performance between the two teaching methods at a 99% confidence level.

Note that we can also calculate the p-value associated with the t-value to determine the exact probability of obtaining the observed results by 
chance. In this case, the p-value is less than 0.01, which supports our decision to reject the null hypothesis.
```

```python
Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean
of 65. Calculate the 90% confidence interval for the true population mean.
ans:
Confidence interval = sample mean ± (critical value) x (standard error)

Where, the critical value is based on the confidence level and the degrees of freedom, and the standard error is calculated as the population standard deviation divided by the square root of the sample size.

First, lets calculate the critical value for a 90% confidence level with 49 degrees of freedom (n-1), using a t-distribution table or calculator:

Critical value = ±1.676

Next, we can calculate the standard error:

Standard error = 8 / sqrt(50) = 1.131

Finally, we can plug in the values to get the confidence interval:

Confidence interval = 65 ± (1.676) x (1.131) = (63.07, 66.93)

Interpretation: We can be 90% confident that the true population mean falls between 63.07 and 66.93.
```

```python
Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average
reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to
determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.
ans:
Null hypothesis (H0): Caffeine has no significant effect on reaction time (μ = 0).
Alternative hypothesis (Ha): Caffeine has a significant effect on reaction time (μ ≠ 0).
We will use a two-tailed t-test at a 90% confidence level.

The t-value can be calculated as:

t = (x̄ - μ) / (s / sqrt(n))

where x̄ is the sample mean (0.25 seconds), μ is the population mean (0), s is the sample standard deviation (0.05 seconds), and n is the sample size (30).

t = (0.25 - 0) / (0.05 / sqrt(30)) = 5.48

Using a t-distribution table or calculator with 29 degrees of freedom (n - 1), we can find the critical values for a two-tailed test at a 90% 
confidence level. The critical values are approximately ±1.699.

Since our calculated t-value (5.48) is greater than the critical value (1.699), we reject the null hypothesis and conclude that caffeine has a
significant effect on reaction time at a 90% confidence level.
```
