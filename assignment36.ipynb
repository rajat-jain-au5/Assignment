```python
# Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
# of 5 using Python. Interpret the results.
# ans:
import scipy.stats as stats

# Sample data parameters
sample_mean = 50
sample_std = 5
sample_size = 100

# Calculate the standard error
std_error = sample_std / (sample_size ** 0.5)

# Calculate the 95% confidence interval
lower_bound, upper_bound = stats.norm.interval(0.95, loc=sample_mean, scale=std_error)

# Print the results
print("The 95% confidence interval is ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```

```python
# Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
# matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
# brown. Use Python to perform the test with a significance level of 0.05.
# ans:
import numpy as np
from scipy.stats import chisquare

# Observed frequencies of M&M colors in a bag
observed_frequencies = np.array([15, 20, 25, 10, 5, 25])

# Expected frequencies based on the given distribution
expected_frequencies = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed_frequencies)

# Perform the chi-square test
chi2_statistic, p_value = chisquare(observed_frequencies, expected_frequencies)

# Set the significance level
alpha = 0.05

# Print the results
print("Chi-square statistic: {:.2f}".format(chi2_statistic))
print("p-value: {:.4f}".format(p_value))
if p_value < 0.05:
    print("Reject the null hypothesis: The distribution of colors of M&Ms in the bag does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of colors of M&Ms in the bag matches the expected distribution.")
```

```python
# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
# smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
# population who smoke.
import numpy as np
import scipy.stats as stats

# Sample size and number of individuals who smoke
n = 500
p = 60/n

# Calculate the standard error and margin of error
se = np.sqrt((p*(1-p))/n)
me = stats.norm.ppf(0.975)*se

# Calculate the confidence interval
lower = p - me
upper = p + me

# Print the results
print("95% confidence interval for the proportion of individuals who smoke: ({:.4f}, {:.4f})".format(lower, upper))
```

```python
# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
# of 12 using Python. Interpret the results.
# ans:
import numpy as np
import scipy.stats as stats

# Sample statistics
sample_mean = 75
sample_std = 12
n = 1  # since we only have one sample

# Calculate the standard error and margin of error
se = sample_std / np.sqrt(n)
me = stats.norm.ppf(0.95) * se

# Calculate the confidence interval
lower = sample_mean - me
upper = sample_mean + me

# Print the results
print("90% confidence interval: ({:.2f}, {:.2f})".format(lower, upper))

    
```

```python
# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
# area corresponding to a chi-square statistic of 15.
# ans:
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define the degrees of freedom
df = 10

# Define the range of the x-axis
x = np.linspace(0, 30, 500)

# Plot the chi-square distribution with 10 degrees of freedom
plt.plot(x, stats.chi2.pdf(x, df), 'k-', lw=2)

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 100)
plt.fill_between(x_shade, stats.chi2.pdf(x_shade, df), color='gray', alpha=0.5)

# Label the axes
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square distribution with 10 degrees of freedom')

# Show the plot
plt.show()
```

```python
Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.
ans:
To calculate the 99% confidence interval for the true proportion of people in the population who prefer Coke, we can use the formula:

CI = (x̄ ± z*(alpha/2) * sqrt(x̄*(1-p-hat)/n))

where:

x̄ is the sample proportion (520/1000 = 0.52)
n is the sample size (1000)
z*(alpha/2) is the critical value of the standard normal distribution at the specified confidence level (99% confidence level corresponds to z*0.005 = 2.576)
alpha is the significance level (1 - confidence level = 0.01)
Plugging in the values, we get:

CI = (0.52 ± 2.576 * sqrt(0.52*(1-0.52)/1000))

CI = (0.484, 0.556)

Therefore, we can be 99% confident that the true proportion of people in the population who prefer Coke is between 0.484 and 0.556
```

```python
Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.
To conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin, we can follow these steps:

1. State the null hypothesis and alternative hypothesis:  
   Null hypothesis: The observed frequencies match the expected frequencies of a fair coin.
   Alternative hypothesis: The observed frequencies do not match the expected frequencies of a fair coin.
2. Determine the expected frequencies for a fair coin:
   The expected frequency for tails is 0.5 (50 times out of 100 flips), and the expected frequency for heads is also 0.5 (50 times out of 100 flips).
3. Calculate the chi-square statistic:
   We can use the formula: chi-square = ∑(observed frequency - expected frequency)^2 / expected frequency
   chi-square = (45 - 50)^2 / 50 + (55 - 50)^2 / 50 = 1
   Note that the degrees of freedom for this test is k-1 = 2-1 = 1, where k is the number of categories (heads and tails).
4. Calculate the p-value:
   We can use a chi-square distribution table  to find the p-value. For this test with 1 degree of freedom and a chi-square statistic of 1, 
   the p-value is approximately 0.317.
5. Make a decision:
   Using a significance level of 0.05, we compare the p-value to the significance level. Since the p-value (0.317) is greater than the significance level (0.05), 
   we fail to reject the null hypothesis.
Therefore, we do not have enough evidence to conclude that the observed frequencies do not match the expected frequencies of a fair coin.
from scipy.stats import chisquare

observed_freq = [45, 55]
expected_freq = [50, 50]

chi2, p_value = chisquare(observed_freq, f_exp=expected_freq)

print("Chi-square statistic:", chi2)
print("p-value:", p_value)
```

```python
Q9. A study was conducted to determine if there is an association between smoking status (smoker or
non-smoker) and lung cancer diagnosis (yes or no). The results are shown in the contingency table below.
Conduct a chi-square test for independence to determine if there is a significant association between
smoking status and lung cancer diagnosis.     
              Lung Cancer: Yes Lung Cancer: No
Smoker            60                     140
Non-smoker        30                     170
Use a significance level of 0.05.   
ans:
Null hypothesis: Smoking status and lung cancer diagnosis are independent.
Alternative hypothesis: Smoking status and lung cancer diagnosis are not independent.

from scipy.stats import chi2_contingency
# contingency table
observed = [[60, 140], [30, 170]]

# chi-square test for independence
chi2, p, dof, expected = chi2_contingency(observed)
# print the results
print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

The chi-square statistic is 16.83, and the p-value is less than 0.05, which means that we reject the null hypothesis. This indicates that there is a 
significant association between smoking status and lung cancer diagnosis. Therefore, we can conclude that smoking status and lung cancer diagnosis are not independent.


```

```python
Q10. A study was conducted to determine if the proportion of people who prefer milk chocolate, dark
chocolate, or white chocolate is different in the U.S. versus the U.K. A random sample of 500 people from
the U.S. and a random sample of 500 people from the U.K. were surveyed. The results are shown in the
contingency table below. Conduct a chi-square test for independence to determine if there is a significant
association between chocolate preference and country of origin.
              Milk Chocolate                  Dark Chocolate    White Chocolate

U.S. (n=500)    200                              150                  150
U.K. (n=500)    225                              175                  100
Use a significance level of 0.01.
ans:
Null hypothesis (H0): There is no significant association between chocolate preference and country of origin.
Alternative hypothesis (Ha): There is a significant association between chocolate preference and country of origin.
The expected frequencies are shown in parentheses.

We can then calculate the chi-square statistic using the formula:

chi-square = Σ[(O - E)2 / E]

where O is the observed frequency and E is the expected frequency.

Calculating the chi-square statistic for this contingency table gives:

chi-square = [(200-240)2/240] + [(150-180)2/180] + [(150-80)2/80] + [(225-210)2/210] + [(175-210)2/210] + [(100-80)2/80]
= 6.75 + 3.75 + 12.5 + 1.25 + 3.75 + 2.5
= 30.00

df = (number of rows - 1) x (number of columns - 1) = 2 x 2 = 4
Using a chi-square distribution table with 4 degrees of freedom and a significance level of 0.01, we find that the critical value is 13.28.

Since the calculated chi-square statistic (30.00) is greater than the critical value (13.28), we reject the null hypothesis and conclude 
that there is a significant association between chocolate preference and country of origin at the 0.01 significance level. 
```

```python
Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.
ans:
To conduct a hypothesis test for the population mean, we need to set up hypotheses:

Null hypothesis (H0): The population mean is not significantly different from 70.
Alternative hypothesis (Ha): The population mean is significantly different from 70.
We can use a significance level of 0.05, which means that we want to be 95% confident in our conclusion.

Since the sample size (n = 30) is greater than 25, we can use a t-test instead of a z-test. The t-test statistic is calculated as:

t = (x̄ - μ) / (s / √n)

where x̄ is the sample mean, μ is the population mean, s is the sample standard deviation, and n is the sample size.

Substituting the values given in the problem, we get:

t = (72 - 70) / (10 / √30) = 1.09

Using a t-distribution table with 29 degrees of freedom (n-1), we find the critical values for a two-tailed test with a significance level of 0.05 are -2.045 and 2.045.
Since the calculated t-statistic (1.09) falls within the acceptance region between -2.045 and 2.045, we fail to reject the null hypothesis.
```
