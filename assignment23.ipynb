```python
Q1. List any five functions of the pandas library with execution.
ans:
Here are five commonly used functions in the Pandas library along with example code that demonstrates their use:

1) read_csv() - This function is used to read a CSV file and return a DataFrame.


import pandas as pd

df = pd.read_csv('data.csv')
print(df.head())

2) groupby() - This function is used to group the data in a DataFrame by one or more columns and perform aggregations on the groups.

# Grouping the DataFrame by a column and calculating the mean of another column
grouped_df = df.groupby('category')['value'].mean()
print(grouped_df)

3) dropna() - This function is used to remove rows or columns from a DataFrame that contain missing values.

# Dropping all rows that contain missing values
cleaned_df = df.dropna()
print(cleaned_df.head())

4) sort_values() - This function is used to sort the rows of a DataFrame by one or more columns.

# Sorting the DataFrame by two columns in ascending order
sorted_df = df.sort_values(['category', 'value'], ascending=[True, True])
print(sorted_df.head())

5) pivot_table() - This function is used to create a summary table of the data in a DataFrame based on one or more columns.

# Creating a pivot table of the DataFrame
pivot_df = pd.pivot_table(df, values='value', index='category', columns='month', aggfunc='mean')
print(pivot_df.head())
```

```python
# Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
# DataFrame with a new index that starts from 1 and increments by 2 for each row.

import pandas as pd

def reindex_df(df):
    new_index = pd.RangeIndex(start=1, step=2, stop=len(df)*2)
    new_df = df.set_index(new_index)
    return new_df

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})
new_df = reindex_df(df)
print(new_df)
```

```python
# Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that
# iterates over the DataFrame and calculates the sum of the first three values in the 'Values' column. The
# function should print the sum to the console.
# For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
# calculate and print the sum of the first three values, which is 60.

import pandas as pd

def calculate_sum(df):
    values_sum = 0
    for i, row in df.iterrows():
        if i < 3:
            values_sum += row['Values']
    print("Sum of first three values in 'Values' column:", values_sum)
    
df = pd.DataFrame({'Values': [10, 20, 30, 40, 50]})
calculate_sum(df)

```

```python
# Q5. How are DataFrame.size() and DataFrame.shape() different?
# ans:
# Both DataFrame.size() and DataFrame.shape() are methods that can be called on a pandas DataFrame object, but they return different information.

# DataFrame.size() returns the total number of elements in the DataFrame, which is equivalent to the product of the number of rows and columns.

# DataFrame.shape() returns a tuple containing the number of rows and columns in the DataFrame. The first element of the tuple is the number of rows and
# the second element is the number of columns.

# Here an example to illustrate the difference between the two methods:

import pandas as pd

data = {'name': ['Alice', 'Bob', 'Charlie', 'David'],
        'age': [25, 30, 35, 40],
        'gender': ['F', 'M', 'M', 'M']}
df = pd.DataFrame(data)

print(df.size)    # Output: 12
print(df.shape)   # Output: (4, 3)
```

```python
# Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email
# addresses in the format 'username@domain.com'. Write a Python function that creates a new column
# 'Username' in df that contains only the username part of each email address.
# The username is the part of the email address that appears before the '@' symbol. For example, if the
# email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
# function should extract the username from each email address and store it in the new 'Username'
# column
# ans:

import pandas as pd

def getUserName(val):
    return val.split('@')[0]

df = pd.DataFrame({'Email': ['john.doe@gmail.com', 'shyam@gmail.com', 'fahim@gmail.com', 'rama@gmail.com']})
df['Username']=df['Email'].map(getUserName)
df
```

```python
# Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects
# all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The
# function should return a new DataFrame that contains only the selected rows.

import pandas as pd

dict1={'A':[3,8,2,6,9],'B':[5,2,9,3,1],'C':[1,7,4,5,2]}
df1 = pd.DataFrame(dict1)
print(df1)
select_rows = df1.loc[(df1['A'] > 5) & (df1['B'] < 10)]
print(select_rows)
```
