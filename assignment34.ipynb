```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
ans:
Estimation statistics is a branch of statistics that deals with estimating population parameters from a sample of data.

Point estimate is a single value that is used to estimate an unknown population parameter based on a sample statistic. 
For example, the sample mean is often used as a point estimate of the population mean. Point estimates are often computed by 
using a sample statistic to estimate the corresponding population parameter.

Interval estimate, on the other hand, is an estimate of a population parameter that includes a range of possible values rather 
than a single point. An interval estimate is usually accompanied by a level of confidence, which indicates the degree of certainty 
that the true population parameter falls within the estimated range. The most common type of interval estimate is the confidence interval. 
Confidence intervals are constructed by using a point estimate and a margin of error that is based on the standard error of the point estimate.

```

```python
Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.
ans:
def estimate_pop_mean(sample_mean, sample_std_dev, sample_size):
    """
    This function estimates the population mean using a sample mean and standard deviation.
    """
    import math
    
    # calculate the standard error
    standard_error = sample_std_dev / math.sqrt(sample_size)
    
    # calculate the margin of error for a 95% confidence interval
    margin_of_error = 1.96 * standard_error
    
    # calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    # return the midpoint of the confidence interval as the estimated population mean
    return (lower_bound + upper_bound) / 2

# example usage
sample_mean = 50
sample_std_dev = 10
sample_size = 100

estimated_pop_mean = estimate_pop_mean(sample_mean, sample_std_dev, sample_size)

print("Estimated population mean:", estimated_pop_mean)
```

```python
Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.
ans:
Hypothesis testing is a statistical method used to test a claim or hypothesis about a population parameter using sample data. It is an essential 
tool in statistics that helps researchers make decisions based on data and evidence.

In hypothesis testing, a researcher formulates two hypotheses, the null hypothesis (H0) and the alternative hypothesis (Ha). The null hypothesis is a
statement of no effect or no difference, while the alternative hypothesis is a statement of the effect or difference that the researcher is interested in.

Hypothesis testing is used to make decisions based on data and evidence. It is commonly used in scientific research, quality control, manufacturing, and 
business decision-making. By using hypothesis testing, researchers can determine whether the observed differences or effects in their data are statistically 
significant, or whether they could have occurred by chance.

The importance of hypothesis testing lies in its ability to provide a rigorous and systematic way of evaluating evidence and making decisions based on data. 
It helps researchers avoid making decisions based on intuition or personal bias, and instead, rely on objective and reproducible evidence. Hypothesis testing 
also allows researchers to quantify the level of uncertainty or error in their conclusions, which is critical for assessing the reliability of their findings.
```

```python
Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.
ans
import numpy as np
from scipy.stats import ttest_ind

# Generate two random samples from two populations with different means and variances
pop1 = np.random.normal(loc=50, scale=10, size=100)
pop2 = np.random.normal(loc=60, scale=10, size=100)

# Calculate the sample means and standard deviations
sample_mean1 = np.mean(pop1)
sample_mean2 = np.mean(pop2)
sample_std_dev1 = np.std(pop1, ddof=1)
sample_std_dev2 = np.std(pop2, ddof=1)

# Conduct a two-sample t-test to compare the means of the two populations
t_stat, p_value = ttest_ind(pop1, pop2)

# Print the results of the hypothesis test
print("Sample mean 1:", sample_mean1)
print("Sample mean 2:", sample_mean2)
print("Sample standard deviation 1:", sample_std_dev1)
print("Sample standard deviation 2:", sample_std_dev2)
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
Q6: What is a null and alternative hypothesis? Give some examples.
ans:
A null hypothesis is a statement that assumes there is no significant difference or relationship between two variables or populations. An alternative 
hypothesis, on the other hand, is a statement that contradicts the null hypothesis and suggests that there is a significant difference or relationship 
between the variables or populations being studied. In hypothesis testing, we aim to either reject or fail to reject the null hypothesis based on the 
evidence available.

Here are some examples of null and alternative hypotheses:

1. Null hypothesis: There is no difference in mean scores between two groups.
   Alternative hypothesis: The mean scores of two groups are significantly different.

2. Null hypothesis: There is no association between smoking and lung cancer.
   Alternative hypothesis: Smoking is significantly associated with an increased risk of lung cancer.

3. Null hypothesis: There is no difference in the proportions of male and female voters who support a particular political candidate.
   Alternative hypothesis: The proportion of male voters who support the political candidate is significantly different from the proportion of female voters who support the political candidate.
```

```python
Q7: Write down the steps involved in hypothesis testing.
ans:
Here are the general steps involved in hypothesis testing:

State the research question and formulate the null and alternative hypotheses.
1. Choose an appropriate statistical test based on the research question and the type of data being analyzed.
2. Determine the level of significance (alpha) and the desired power of the test.
3. Collect and analyze the data, calculate the test statistic, and obtain the p-value.
4. Compare the p-value to the level of significance (alpha) to determine whether to reject or fail to reject the null hypothesis.
5. Interpret the results of the hypothesis test in the context of the research question and draw conclusions.
6. Report the findings, including the results of the hypothesis test, any assumptions made, and the limitations of the study.
```

```python
Q8. Define p-value and explain its significance in hypothesis testing.
ans:
In hypothesis testing, the p-value is the probability of obtaining a test statistic as extreme or more extreme than the one observed, assuming that the
null hypothesis is true. In other words, the p-value is the probability of observing a result as extreme as, or more extreme than, the one obtained if 
the null hypothesis were true.

The p-value plays a significant role in hypothesis testing because it is used to determine whether to reject or fail to reject the null hypothesis. The 
general rule is that if the p-value is less than or equal to the level of significance (alpha), then the null hypothesis is rejected, and the alternative 
hypothesis is accepted. On the other hand, if the p-value is greater than the level of significance, then the null hypothesis is not rejected, and the
alternative hypothesis is not supported.

For example, if the level of significance is set at 0.05, and the p-value obtained from the hypothesis test is 0.02, then the null hypothesis would be 
rejected, and the alternative hypothesis would be accepted. This would indicate that the result observed is statistically significant and unlikely to occur 
by chance alone, assuming that the null hypothesis is true.
```

```python
Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.
ans
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

# Set the degrees of freedom
df = 10

# Generate the x values for the plot
x = np.linspace(-4, 4, 500)
# Generate the y values for the t-distribution
y = t.pdf(x, df)

# Create the plot
fig, ax = plt.subplots()
ax.plot(x, y)

# Add a vertical line at 0
ax.axvline(x=0, color='grey', linestyle='--')

# Set the axis labels and title
ax.set_xlabel('x')
ax.set_ylabel('Probability density')
ax.set_title(f'Student\'s t-Distribution (df={df})')

# Show the plot
plt.show()
```

```python
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.
ans:
import numpy as np
from scipy.stats import ttest_ind

# Generate two random samples with equal size
sample1 = np.random.normal(loc=10, scale=2, size=50)
sample2 = np.random.normal(loc=12, scale=2, size=50)

# Set the null hypothesis that the population means are equal
null_hypothesis = 'The population means are equal'

# Calculate the t-test and print the results
t_statistic, p_value = ttest_ind(sample1, sample2, equal_var=True)
print(f'Two-Sample t-test for Independent Samples\nNull Hypothesis: {null_hypothesis}\n')
print(f'T-Statistic: {t_statistic:.2f}')
print(f'P-Value: {p_value:.4f}')
```

```python
Q11: What is Students t distribution? When to use the t-Distribution.
ans:
Students t-distribution is a probability distribution that is used to estimate the population mean when the sample size is small and/or the population standard deviation 
is unknown. 
The t-distribution is similar to the standard normal distribution, but has heavier tails, which makes it more appropriate for small sample sizes. As the sample size increases, 
the t-distribution approaches the standard normal distribution.

The t-distribution is used in hypothesis testing, particularly when the sample size is small or the population standard deviation is unknown. It is used to
test hypotheses about population means, such as whether two sample means are different, or whether a sample mean is significantly different from a 
hypothesized value. The t-distribution is also used to construct confidence intervals for population means.
```

```python
Q12: What is t-statistic? State the formula for t-statistic.
ans:
The t-statistic is a measure of how many standard errors the sample mean is away from the hypothesized population mean. It is used to test 
hypotheses about population means when the population standard deviation is unknown and the sample size is small.

The formula for the t-statistic is:

t = (x - ) / (s / sqrt(n))

where:

x is the sample mean
 is the hypothesized population mean
s is the sample standard deviation
n is the sample size
The t-statistic measures the difference between the sample mean and the hypothesized population mean, relative to the variability in the sample.
If the absolute value of the t-statistic is large, then it indicates that the difference between the sample mean and the hypothesized population mean 
is unlikely to have occurred by chance, and we can reject the null hypothesis.
The t-distribution is used to calculate the p-value for the t-statistic, which indicates the probability of obtaining a t-statistic as extreme or more extreme
than the one observed, assuming the null hypothesis is true. If the p-value is less than the significance level (usually 0.05), then we reject the null
hypothesis and conclude that the sample mean is significantly different from the hypothesized population mean.
```

```python
Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.
ans:
The test statistic for a one-sample t-test is given by:

t = (x - ) / (s / sqrt(n))

where:

x is the sample mean weight (4.8 pounds)
 is the hypothesized population mean weight (5 pounds)
s is the sample standard deviation (0.5 pounds)
n is the sample size (25)
Plugging in the values, we get:

t = (4.8 - 5) / (0.5 / sqrt(25))
t = -2

From a t-distribution table we can find that the p-value for a one-tailed test with 24 degrees of freedom and t = -2 is approximately 0.027.

Since the p-value is less than 0.01, we can reject the null hypothesis and conclude that there is evidence that the true mean weight of the products is 
less than 5 pounds.
```

```python
Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.
ans:
Null hypothesis: The population means for the two groups are equal.
Alternative hypothesis: The population means for the two groups are not equal.
We can use the following formula to calculate the t-statistic:

t = (x1 - x2) / sqrt(s1^2/n1 + s2^2/n2)
Using the given data:

x1 = 80, s1 = 10, n1 = 30
x2 = 75, s2 = 8, n2 = 40

dof = (s1^2/n1 + s2^2/n2)^2 / ((s1^2/n1)^2 / (n1-1) + (s2^2/n2)^2 / (n2-1))

dof = (10^2/30 + 8^2/40)^2 / ((10^2/30)^2 / 29 + (8^2/40)^2 / 39) = 67.57 

For a two-tailed test, the critical t-value is approximately 2.66.

t = (80 - 75) / sqrt(10^2/30 + 8^2/40) = 2.29

Since the calculated t-value of 2.29 is less than the critical t-value of 2.66, we fail to reject the null hypothesis. Therefore, we can conclude that
there is not enough evidence to suggest that the population means for the two groups are different at a significance level of 0.01.
```
