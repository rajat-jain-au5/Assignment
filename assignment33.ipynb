```python
Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
ans:
The Cumulative Distribution Function (CDF) is a function that describes the probability that a random variable X takes on a value less than or equal to a 
given value x. In other words, it gives the cumulative probability distribution of the random variable.

The CDF is defined as:

F(x) = P(X <= x)

where X is a random variable and x is a specific value. The CDF gives the probability of the variable being less than or equal to x.

For example, consider the toss of a coin. The CDF of this variable would give the probability that the number of heads obtained in a certain number of coin 
tosses is less than or equal to a given value. Suppose we want to calculate the probability of obtaining two or fewer heads in three tosses of a fair coin. 
The possible outcomes are:

HHH, HHT, HTH, THH, HTT, THT, TTH, TTT

Out of these, there are three outcomes where we obtain two or fewer heads (namely, HHH, HHT, and HTH). Therefore, the probability of obtaining two or 
fewer heads is 3/8. The CDF for this example would give the probability of obtaining two or fewer heads in any number of tosses.
CDF is used to calculate various probabilities and statistics of a random variable. Some of the common statistics that can be obtained from the CDF 
include the mean, median, mode, variance, and standard deviation of the distribution. CDFs are also used to calculate the percentiles of a distribution. 
```

```python
Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.
ans:
The normal distribution is widely used as a model in situations where the distribution of a variable is symmetric and bell-shaped. 
Some common examples of situations where the normal distribution might be used as a model include:

1. Heights and weights of people in a population
2. Scores on standardized tests, such as the SAT or ACT
3. IQ scores in a population
4. The length of time it takes to complete a task
5. The distribution of errors in measurements
The parameters of the normal distribution are the mean (μ) and the standard deviation (σ). The mean determines the location of the center of the distribution, 
while the standard deviation determines the spread of the distribution. The shape of the distribution is determined by the values of μ and σ.

When μ is small, the distribution is shifted to the left, and when μ is large, the distribution is shifted to the right. When σ is small, the distribution is
tightly clustered around the mean, while when σ is large, the distribution is more spread out.

In a normal distribution, about 68% of the observations fall within one standard deviation of the mean, about 95% fall within two standard deviations, 
and about 99.7% fall within three standard deviations. This is known as the empirical rule or the 68-95-99.7 rule.
```

```python
Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?
ans:
The Bernoulli distribution is a probability distribution that models the outcome of a single binary event, where the event can have one of two possible outcomes, 
typically denoted as success or failure. The Bernoulli distribution is named after Swiss mathematician Jacob Bernoulli.

1. The Bernoulli distribution is characterized by a single parameter, p, which represents the probability of a success in a single trial. The probability 
mass function of the Bernoulli distribution is:

P(X = 1) = p
P(X = 0) = 1 - p

where X is the random variable that represents the outcome of a single trial.

A real-life example of the Bernoulli distribution is flipping a coin. If we consider a successful outcome as obtaining a head and a failure outcome 
as obtaining a tail, then the probability of obtaining a head in a single flip of a fair coin is 0.5. Thus, the outcome of flipping a fair coin can
be modeled by a Bernoulli distribution with p = 0.5.

The difference between the Bernoulli distribution and the binomial distribution is that the Bernoulli distribution models the outcome of a single trial, 
while the binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.

2. In the binomial distribution, we have a fixed number of independent trials, denoted as n, and the probability of success in each trial, denoted as p, is 
constant across all trials. The random variable in the binomial distribution is the number of successes in the n trials, denoted as X.

The probability mass function of the binomial distribution is:

P(X = k) = (n choose k) * p^k * (1 - p)^(n-k)

where (n choose k) is the binomial coefficient, which represents the number of ways to choose k successes out of n trials.

For example, if we flip a fair coin 10 times and count the number of heads obtained, the number of heads obtained can be modeled by a binomial 
distribution with n = 10 and p = 0.5. The probability of obtaining exactly 5 heads in 10 flips of a fair coin can be calculated using the binomial probability mass function.
    
```

```python
Q7: Explain uniform Distribution with an example.
ans:
Uniform distribution is a probability distribution where all values in a given range are equally likely to occur. This means that the probability of observing 
any value within the range is the same.

The probability density function (PDF) of a continuous uniform distribution is given by:

f(x) = 1 / (b - a) if a <= x <= b
0 otherwise

where a and b are the minimum and maximum values of the range, respectively.

For example, consider the probability of randomly selecting a number between 1 and 6, inclusive, from a fair six-sided die. Since each of the six numbers is 
equally likely to be selected, we can model this using a continuous uniform distribution on the range [1, 6]. In this case, the minimum value (a) is 1 and the 
maximum value (b) is 6. The PDF of the uniform distribution for this case would be:

f(x) = 1/6 if 1 <= x <= 6
0 otherwise

The probability of observing any value within the range [1, 6] is the same, which is 1/6. For example, the probability of rolling a 3 is 1/6, the probability 
of rolling a 5 is 1/6, and so on.

Uniform distribution is often used to model situations where all outcomes are equally likely, such as in the case of rolling a fair die, selecting a random card 
from a deck of cards, or selecting a random number from a given range.
```

```python
Q8: What is the z score? State the importance of the z score.
ans:
A z-score (or standard score) is a statistical measurement that indicates how many standard deviations a data point is from the mean of a dataset. It is calculated using the formula:

z = (x - μ) / σ

where x is the observed value, μ is the mean of the dataset, and σ is the standard deviation of the dataset.

The importance of the z-score is that it allows us to compare values from different datasets that may have different means and standard deviations.
By standardizing the data using z-scores, we can determine how unusual or typical a particular value is relative to the rest of the data.

Z-scores are also used to determine the probability of observing a particular value or set of values in a normal distribution. By converting observed values to z-scores 
and looking up the corresponding area under the standard normal distribution curve, we can determine the probability of observing values that are more extreme than the 
observed values. This is useful for hypothesis testing and statistical inference.
```

```python
Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
ans:
The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that the sampling distribution of the mean of a large number of 
independent and identically distributed (i.i.d) random variables will approximate a normal distribution, regardless of the underlying distribution of
the individual variables. In other words, as the sample size increases, the distribution of the sample means will become more normal, regardless of the 
distribution of the individual data points.

1. The CLT has several significant implications in statistics and data analysis, including:

2. It allows us to make inferences about population parameters based on samples, even when the population distribution is unknown or not normally distributed. 
By taking a large enough sample, the CLT ensures that the sample means will be normally distributed, allowing us to use common statistical methods and tests 
that assume normality.

3. It provides a justification for the use of the mean as a summary statistic for large datasets, as the mean of a large sample will be approximately normally
distributed, even if the underlying data is not normally distributed.

4. It helps to explain why normal distributions are so common in nature and in statistical data, as they can arise from the aggregation of many small, random effects.

It is a foundation for many advanced statistical techniques, including regression analysis, hypothesis testing, and confidence intervals.
```

```python
Q10: State the assumptions of the Central Limit Theorem.
ans:
The Central Limit Theorem (CLT) has some assumptions that must be met for it to be applicable. These assumptions include:

1. Independence: The sample data must be collected independently of one another. This means that the selection of one data point should not influence 
the selection of another data point.

2. Randomness: The sample data should be selected randomly from the population of interest. This means that each data point in the population should have an 
equal chance of being included in the sample.

3. Sample Size: The sample size should be sufficiently large. While there is no exact threshold for the sample size, it is generally recommended that the sample 
size be at least 30 or more to apply the CLT.

4. Finite Variance: The population distribution should have a finite variance. This means that the distribution should not have extreme values that are too far from the mean.
```
