```python
Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
ans: The ArithmeticError is a built-in class in Python that serves as a base class for all arithmetic-related exceptions. 
Here are two common errors that are defined in the ArithmeticError class:

1.ZeroDivisionError: This error is raised when a division by zero occurs. For example, dividing any number by zero will result in this error.
Example:
# Division by zero
a = 10
b = 0
c = a/b  # ZeroDivisionError: division by zero

2.OverflowError: This error is raised when the result of an arithmetic operation is too large to be represented by the available memory.
Example:
# Overflow error
a = 2**1000
b = a**a  # OverflowError: (34, 'Numerical result out of range')
```

```python
Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
ans: The LookupError class is used as a base class for exceptions that occur when an item is not found in a collection, such as a list or dictionary. 
     Here are two common errors that are defined as subclasses of LookupError:
    1. KeyError: This error is raised when we try to access a non-existent key in a dictionary.
       for example
try :
    d = {"key" :"sudh" , 1 : [2,3,4,5]}
    print(d["key2"])
except KeyError as e : 
    print(e)      
    2. IndexError: This error is raised when we try to access an element from a list or tuple that is outside the bounds of the collection.
    for example :
try :
    l = [2,3,4,5]
    print(l[6])
except IndexError as e :
    print(e)
```

```python
Q5. Explain ImportError. What is ModuleNotFoundError?
ans: ImportError is an exception that is raised when a module or package cannot be imported by Python. This can occur when the module is not found, or when there is 
     an error in the code of the module.
     ModuleNotFoundError is a subclass of ImportError that is raised when a module cannot be found by Python. This error occurs when the specified module cannot be 
     located in the Python path.
# Example of ImportError
try:
    import 'sudh  # sudh does not exist
except ImportError as e:
    print("ImportError: ", e)

# Output: ImportError: No module named 'my_module'

# Example of ModuleNotFoundError
try:
    import my_module  # my_module does not exist
except ModuleNotFoundError as e:
    print("ModuleNotFoundError: ", e)

# Output: ModuleNotFoundError: No module named 'my_module'
```
